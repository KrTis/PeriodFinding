```python
stars = pd.read_csv('data/query_1610017978.44685i.txt',skiprows=6,sep='\t')
stars['field']=stars['# ID'].str.split('-').apply(lambda x:x[1])
stars['Iband']=stars['# ID'].apply(lambda x:'data/ogle/I/'+x+'.dat')
stars['Vband']=stars['# ID'].apply(lambda x:'data/ogle/V/'+x+'.dat')
stars = stars[stars['field']=='BLG']
Analysis = pd.DataFrame()
Nreps = 100
Percents = [7,10,15,50,70]
def p16(x):
    
    return np.percentile(x,16)
def p84(x):
    return np.percentile(x,84)
def med_sig(x):
    return (np.percentile(x,75)-np.percentile(x,25))/1.349
def unexplained_variance(x):
    return np.sum((x-np.median(x))**2)/x.size-med_sig(x)**2
def Outliers1(x):
    N = x.size
    s = med_sig(x)
    m = np.median(x)
    return (np.sum(x<m-s)+ np.sum(x>m+s))/N
def Outliers3(x):
    N = x.size
    s = med_sig(x)
    m = np.median(x)
    return (np.sum(x<m-3*s)+ np.sum(x>m+3*s))/N
stats = ['mean','median','std','sem',p16,p84,med_sig,unexplained_variance,Outliers1,Outliers3]
Stats = ['mean','median','std','sem','p16','p84','med_sig','unexplained_variance', 'Outliers1','Outliers3']
cluster = SSHCluster(
                ["dodo1",'dodo2','dodo3','dodo4']+["dodo5",'dodo6'],
                connect_options={"known_hosts": None},
                scheduler_options={"port": 0, "dashboard_address": ":8797"}
            )
for i in stars.index[0:10]:
        print(stars.loc[i,'# ID'],end='\t')
        inds = []
        try:
            Iband = pd.read_csv(stars.loc[i,'Iband'],names=['t','mag','magerr'],sep=' ')
            Iband['filt'] = 'I'
            inds.append(0)
        except:
            print('No I')
            Iband = None
        try:
            Vband = pd.read_csv(stars.loc[i,'Vband'],names=['t','mag','magerr'],sep=' ')
            Vband['filt']= 'V'
            inds.append(1)
        except:
            print('No V')
            Vband=None
        if len(inds)==2:
            D=pd.concat([Iband,Vband])
        elif len(inds)==1:
            D = [Iband,Vband][inds[0]]
        else:
            continue
        file = 'data/ogle/sources/'+stars.loc[i,'# ID']+'.dat'
        D.to_csv(file)
        N_I = (D['filt']=='I').count()
        N_V = (D['filt']=='V').count()
        Nmax = N_I+ N_V
        Sim = MCSimulation('data/ogle/sources/', stars.loc[i,'P_1'], file)
        #sims=np.unique(np.round(Nmax*Percents).astype(int))
        
        
        Dict = {'Name':stars.loc[i,'# ID'],'OGLE period':stars.loc[i,'P_1'],'LS period':Sim.best_fitting[1],'N_I':N_I,'N_V':N_V}
        #for Per in Percents:
        ok = False
        for _ in range(3):
                try:
                    Sim.produce_bootstrap(Percents,Nreps)
                    #Sim.produce_bootstrap([Per],Nreps)
                    Sim.run_simulation('fast',cluster=cluster)
                    Stat = Sim.Simulated_periods.groupby('N').agg({'P':stats})
                    ok = True
                except:
                    print(Percents,'Fail',end='\t')
                if ok:
                    for ind in Stat.index:
                        for st in Stats:
                            Dict[f'{ind} '+st] = Stat.loc[ind,('P',st)]
                    break
        print(ok)
        Analysis = Analysis.append(Dict, ignore_index=True)
        Analysis.to_csv('Manystars.csv')
```

```python
Percents = [7,10,15,50]
fig, ax = plt.subplots(ncols=3, nrows=2, constrained_layout=True,figsize=(12,8),
                             gridspec_kw=dict(width_ratios=[1,1,1], height_ratios=[1,1]),dpi=150)
ax = ax.flatten()
for s in Percents[::-1]:
    ax[0].hist((Analysis[f'{s} median']-Analysis['OGLE period'])/Analysis[f'{s} med_sig'],5,label=f'N={s}',histtype='step',)
    ax[1].scatter(Analysis['OGLE period'],Analysis[f'{s} med_sig']/Analysis['OGLE period'],label=f'N={s}')
    ax[2].hist(abs(Analysis[f'{s} median']-Analysis['OGLE period'])/Analysis['OGLE period'],np.logspace(-5,1,10),histtype='step',label=f'N={s}')
    ax[3].hist(np.sqrt(np.abs(Analysis[f'{s} unexplained_variance']/Analysis['OGLE period']**2)),np.logspace(-5,2,20),histtype='step',label=f'N={s}')
    ax[4].scatter(Analysis['OGLE period'],Analysis[f'{s} Outliers1'],label=f'N={s}')
    ax[5].scatter(Analysis['OGLE period'],Analysis[f'{s} Outliers3'],label=f'N={s}')

for _, _x, _y, _xs, _ys in zip(ax,[r'$\frac{P-P_{OGLE}}{\sigma_{IQR}}$',r'$P_{OGLE}$[d]',r'$\left|\frac{P_{median}-P_{OGLE}}{P_{OGLE}}\right|$',r'$\sqrt{\sigma^2-\sigma^2_{IQR}}/P_{OGLE}$']+2*[r'$P_{OGLE}$[d]'],
                     [r'$P_{median}/P_{OGLE}$',r'$\sigma_{IQR}/P_{OGLE}$',"","", r'$1\sigma_{IQR}$ outliers', r'$3\sigma_{IQR}$ outliers'], 2*['linear']+2*['log']+2*['linear'],['linear','log']+4*['linear']):
    _.set_xlabel(_x)
    _.set_ylabel(_y)
    _.legend()
    _.set_xscale(_xs)
    _.set_yscale(_ys)
ax[4].axhline(1-0.6827,color='k',ls='dashed')
ax[5].axhline(1-0.9973,color='k',ls='dashed')
```
